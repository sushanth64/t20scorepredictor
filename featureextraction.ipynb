```python
import os
import pickle

# Directory path
directory_path = '/kaggle/input/dataset-level2'

# List all files in the directory
files = os.listdir(directory_path)
print("Files in directory:", files)

# Choose the specific file to load
file_name = 'dataset_level2 (1).pkl'
file_path = os.path.join(directory_path, file_name)

# Load the pickle file
with open(file_path, 'rb') as f:
    df = pickle.load(f)

# Now you can work with the loaded DataFrame `df`
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
import pandas as pd

# Group the DataFrame by 'match_id'
groups = df.groupby('match_id')

# Initialize an empty list to hold the rolling sums
all_rolling_sums = []

# Iterate over each group (match)
for name, group in groups:
    # Calculate the rolling sum of 'runs' with a window of 30
    rolling_sums = group['runs'].rolling(window=30).sum()
    
    # Append the rolling sums to the list, preserving the original DataFrame index
    all_rolling_sums.extend(rolling_sums)

# Ensure the length of all_rolling_sums matches the DataFrame length
assert len(all_rolling_sums) == len(df), "Length of rolling sums does not match DataFrame length."

# Add the rolling sums as a new column to the DataFrame
df['last_five'] = all_rolling_sums

# Output the DataFrame with the new column
print(df.head())
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
